```python
from tkinter import *
import tkinter as tk

master = Tk()
master.title(" TO-DO-LIST ")
master.config(bg = "blue")

frame1 = Frame(master, bg = "black", bd = 8, relief = "sunken")
frame1.grid(row = 2, column = 1)
def add_task():
    task = entry.get() 
    if task:
        listbox.insert(END, task)
        entry.delete(0, END)

def mark_completed():
    selected_task_index = listbox.curselection()
    if selected_task_index:
        completed_task = listbox.get(selected_task_index)
        listbox.delete(selected_task_index)
        completed_listbox.insert(END, completed_task)

def delete_task():
    selected_task_index = listbox.curselection()
    if selected_task_index:
        listbox.delete(selected_task_index)
        
        
Label(master, text="Smartech To-do-list", bg = "grey", fg = "white",font = ("Algerian",20)).grid(row=0, column = 1, columnspan =2)
Label(master, text="Today's Task", bg = "black", fg = "white",font = ("Algerian",20)).grid(row=3, column = 1, columnspan =2)
Label(master, text="Completed Tasks", bg = "grey", fg = "white",font = ("Algerian",20)).grid(row=5, column = 1, columnspan =2)        
entry = Entry(master, width = 50, bd =8, relief = "sunken")
entry.grid(row = 1, column = 1)

listbox = Listbox(master, bg = "white", bd = 6, relief = "sunken")
listbox.grid(row =4, column=1, columnspan = 2, ipadx = 60, ipady = 10)

completed_listbox = Listbox(master, bg = "pink", bd = 6, relief = "raised")
completed_listbox.grid(row=6, column=1, columnspan = 2, ipadx=60, ipady = 10)

Button(frame1, text="Add Task", command=add_task, fg = "black", relief = "raised", bd = 6).grid(row=2, column=0, padx=2, pady=3)
Button(frame1, text="Mark Completed", command=mark_completed, fg = "green", relief = "raised", bd = 6).grid(row=2, column=1, padx=2, pady=3)
Button(frame1, text="Delete Task", command=delete_task, fg = "red", relief = "raised", bd = 6).grid(row=2, column=2, padx=2, pady=3)

master.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
from tkinter import simpledialog
from tkinter import font

master = tk.Tk()
master.title("Rich Text Note-Taking App")


# Creating a Text widget for note-taking
text = tk.Text(master, wrap=tk.WORD, bg = "yellow")
text.pack(fill=tk.BOTH, expand=True)

# Creating a toolbar with formatting buttons
toolbar = tk.Frame(master, bg = "pink")
toolbar.pack(fill=tk.X)


def set_bold():
    current_tags = text.tag_names("sel.first")
    if "bold" in current_tags:
        text.tag_remove("bold", "sel.first", "sel.last")
    else:
        text.tag_add("bold", "sel.first", "sel.last")
        bold_font = font.nametofont(text.tag_cget("bold", "font"))
        bold_font.configure(weight="bold")
        text.tag_configure("bold", font=bold_font)

def set_italic():
    current_tags = text.tag_names("sel.first")
    if "italic" in current_tags:
        text.tag_remove("italic", "sel.first", "sel.last")
    else:
        text.tag_add("italic", "sel.first", "sel.last")
        italic_font = font.nametofont(text.tag_cget("italic", "font"))
        italic_font.configure(slant="italic")
        text.tag_configure("italic", font=italic_font)

def set_underline():
    current_tags = text.tag_names("sel.first")
    if "underline" in current_tags:
        text.tag_remove("underline", "sel.first", "sel.last")
    else:
        text.tag_add("underline", "sel.first", "sel.last")
        text.tag_configure("underline", underline=True)

def change_font(font_family):
    text.tag_add(font_family, "sel.first", "sel.last")
    text.tag_configure(font_family, font=(font_family, text.tag_cget("normal", "size")))

def change_color(color):
    text.tag_add(color, "sel.first", "sel.last")
    text.tag_configure(color, foreground=color)

def insert_bullet():
    text.insert("insert", "â€¢ ", "bullet")

def save_file():
    note_text = text.get("1.0", "end-1c")
    filename = simpledialog.asksaveasfilename(defaultextension=".txt", filetypes=[("Text Files", "*.txt")])
    if filename:
        with open(filename, "w") as file:
            file.write(note_text)
        messagebox.showinfo("Info", "File saved successfully.")

def open_file():
    filename = simpledialog.askopenfilename(filetypes=[("Text Files", "*.txt")])
    if filename:
        with open(filename, "r") as file:
            note_text = file.read()
            text.delete("1.0", "end")
            text.insert("1.0", note_text)

bold_button = tk.Button(toolbar, text="Bold", command=set_bold)
bold_button.pack(side=tk.LEFT)

italic_button = tk.Button(toolbar, text="Italic", command=set_italic)
italic_button.pack(side=tk.LEFT)

underline_button = tk.Button(toolbar, text="Underline", command=set_underline)
underline_button.pack(side=tk.LEFT)

font_family_var = tk.StringVar()
font_family_var.set("Arial")
font_family_dropdown = tk.OptionMenu(toolbar, font_family_var, "Arial", "Times New Roman", "Courier New", command=lambda _: change_font(font_family_var.get()))
font_family_dropdown.pack(side=tk.LEFT)

color_var = tk.StringVar()
color_var.set("black")
color_dropdown = tk.OptionMenu(toolbar, color_var, "black", "red", "blue", "green", command=lambda _: change_color(color_var.get()))
color_dropdown.pack(side=tk.LEFT)

bullet_button = tk.Button(toolbar, text="Bullet Point", command=insert_bullet)
bullet_button.pack(side=tk.LEFT)

# Creating a menu for saving and loading notes
menu = tk.Menu(master)
master.config(menu=menu)
file_menu = tk.Menu(menu)
menu.add_cascade(label="File", menu=file_menu)
file_menu.add_command(label="Save", command=save_file)
file_menu.add_command(label="Open", command=open_file)
file_menu.add_separator()
file_menu.add_command(label="Exit", command=root.quit)

master.mainloop()
```
