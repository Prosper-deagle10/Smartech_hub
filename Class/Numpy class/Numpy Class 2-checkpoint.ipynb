# Correlation

```python
np.random.seed(30)
age = np.random.randint(16,70,15)
weight = np.random.randint(40,170,15)
print(age)
print(weight)
```

```python
import matplotlib.pyplot as plt
plt.scatter(age,weight)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(age,weight)

plt.title('Scatter Plot Example')
plt.xlabel('Age')
plt.ylabel('Weight')

plt.show()
```

```python
np.corrcoef(age,weight)
```

```python
plt.scatter(age,weight, s = 100, alpha = 0.5)

plt.title('Scatter Plot Example')
plt.xlabel('Age')
plt.ylabel('Weight')

plt.show()
```

```python
np.mean(age) 
```

```python
np.std(age)
```

```python
plt.boxplot(age)
plt.show()
```

```python
plt.boxplot(weight)
plt.show()
```

```python
mean = age.mean()
std = age.std()
z = (age -mean)/std
print(z)
```

```python
mod_age = age[:]
print(mod_age)
```

```python
mod_age[13] = 700
print(mod_age)
```

```python
np.std(mod_age)
```

```python
plt.boxplot(mod_age)
plt.show()
```
