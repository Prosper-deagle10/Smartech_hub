# Correlation

```python
np.random.seed(30)
age = np.random.randint(16,70,15)
weight = np.random.randint(40,170,15)
print(age)
print(weight)
```

```python
import matplotlib.pyplot as plt
plt.scatter(age,weight)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(age,weight)

plt.title('Scatter Plot Example')
plt.xlabel('Age')
plt.ylabel('Weight')

plt.show()
```

```python
np.corrcoef(age,weight)
```

```python
plt.scatter(age,weight, s = 100, alpha = 0.5, color = "red")

plt.title('Scatter Plot Example')
plt.xlabel('Age')
plt.ylabel('Weight')

plt.show()
```

```python
plt.boxplot(weight)
plt.show()
```

```python
mean = age.mean()
std = age.std()
z = (age -mean)/std
print(z)
```

# stack plot

```python
import matplotlib.pyplot as plt

# Example data
time = [1, 2, 3, 4, 5]
variable1 = [2, 3, 5, 7, 10]
variable2 = [1, 2, 4, 5, 8]

# Stack plot
plt.stackplot(time, variable1, variable2, labels=['Variable 1', 'Variable 2'])
plt.legend(loc='upper left')
plt.xlabel('Time')
plt.ylabel('Cumulative Value')
plt.title('Stack Plot Example')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate example data
time = np.arange(1, 6)
variable1 = np.array([2, 3, 5, 7, 10])
variable2 = np.array([1, 2, 4, 5, 8])

# Calculate cumulative values
cumulative1 = np.cumsum(variable1)
cumulative2 = np.cumsum(variable2)

# Stack plot using NumPy
plt.stackplot(time, cumulative1, cumulative2, labels=['Variable 1', 'Variable 2'])
plt.legend(loc='upper left')
plt.xlabel('Time')
plt.ylabel('Cumulative Value')
plt.title('Stack Plot Example with NumPy')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating subplots
fig, axes = plt.subplots(2, 1, sharex=True)

# Plotting in subplots
axes[0].plot(x, y1, label='sin(x)')
axes[0].set_title('Sine Function')

axes[1].plot(x, y2, label='cos(x)', color='orange')
axes[1].set_title('Cosine Function')

# Adding legends and adjusting layout
for ax in axes:
    ax.legend()
plt.tight_layout()
plt.show()
```
