## Write a program that will create a histogram, scatter plot, and box plot using matplotlib.

```python
# 1. Histogram. 
# A histogram is a graph showing frequency distributions. It is a graph showing the number of 
# observations within each given interval
import numpy as np
import matplotlib.pyplot as plt

ages = np.random.randint(10,50, 30)
print(ages)
```

```python
# To create the freuency distribution of ages. We use the command:
plt.hist(ages)
plt.show()
```

```python
age = np.random.randint(10,50,20)
print(age)
```

```python
plt.hist(age)
plt.show()
# the plot will show the distribution of ages based on the high frequency. 46 appear thrice so thats the highest.
```

```python
# 2. Scatter plot
# A scatter plot is a type of data visualization that displays individual data points on a 
# two-dimensional graph. Each point represents the values of two variables and is plotted along the x and y axes. 
# Scatter plots are useful for visualizing the relationship between two continuous variables and identifying patterns, 
# clusters, or trends in the data.

x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 12, 5, 8, 6])
plt.scatter(x,y)
plt.show()
# In this case, when x is 1, y = 10, when x = 2, y = 12, and so on.
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 12, 5, 8, 6])
plt.scatter(x,y)

plt.title('Scatter Plot Example')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 12, 5, 8, 6])
plt.scatter(x,y)

# to compare two variables 
x = np.array([2, 4, 6, 7, 8])
y = np.array([12, 10, 12, 5, 10])

plt.scatter(x,y)
plt.title('Scatter Plot Example')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```

```python
# 3. Box plot
# A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of the
# distribution of a dataset. It displays key statistical measures, such as the median, quartiles, and potential outliers.
# Matplotlib provides a simple way to create box plots using the boxplot function.

import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
data_set1 = np.random.normal(0, 1, 100)
data_set2 = np.random.normal(3, 1, 100)
data_set3 = np.random.normal(7, 2, 100)

# Combine data sets
data = [data_set1, data_set2, data_set3]

# Create a box plot
plt.boxplot(data, vert=True, patch_artist=True, notch=True, sym='r+')

# Customize the plot
plt.title('Box Plot Example')
plt.xlabel('Data Sets')
plt.ylabel('Values')
plt.grid(True)

# Display the plot
plt.show()

# In this example, three datasets are combined, and a box plot is created. The boxplot function is used with 
# various parameters to customize the appearance of the plot. 
```

## Study universal function in numpy and aggregate functions, give examples, and show the differences btwn Aggregate functions and Universal functions.

# Universal Functions

```python
# In NumPy, universal functions (ufuncs) are functions that operate element-wise on arrays, 
# performing element-wise operations across arrays of any size and shape. 
# They are a core component of NumPy and provide a fast and efficient way to perform operations on entire 
# arrays without the need for explicit looping. 
# Ufuncs are a key reason why NumPy is so powerful and efficient for numerical computing.

# Examples of Universal functions (ufuncs)
# 1 Element-wise operations - Ufuncs perform element-wise operations, applying a function to each element in an array.


arr = np.array([1, 2, 3, 4])
result = np.add(arr, 2)
print(result)
```

```python
# 2. Array Broadcasting
# Ufuncs can automatically broadcast (extend) smaller arrays to match the shape of larger arrays,
# making it easy to perform operations on arrays of different shapes.
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([10, 20, 30, 40])
result = np.add(arr1, arr2)
print(result)
```

# Aggregate functions

```python
# Aggregate functions, also known as reduction functions, operate on an entire array and produce a single result.

# Example: sum of all elements
arr = np.array([1, 2, 3, 4])
result = np.sum(arr)
print(result)
```
