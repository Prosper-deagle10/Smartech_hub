# Stack/Area plot

```python
# A stack plot, also known as a stacked area plot or area chart, is a type of data visualization that displays 
# the contribution of different variables to a total over time or any other continuous dimension. 
# It is particularly useful for illustrating the cumulative values of multiple variables 
# and their evolution over a certain period.

import numpy as np
import matplotlib.pyplot as plt

# Generate example data
time = np.arange(1, 6)
variable1 = np.array([2, 3, 5, 7, 10])
variable2 = np.array([1, 2, 4, 5, 8])

# Calculate cumulative values
cumulative1 = np.cumsum(variable1)
cumulative2 = np.cumsum(variable2)

# Stack plot using NumPy
plt.stackplot(time, cumulative1, cumulative2, labels=['Variable 1', 'Variable 2'])
plt.legend(loc='upper left')
plt.xlabel('Time')
plt.ylabel('Cumulative Value')
plt.title('Stack Plot Example with NumPy')
plt.show()

# In this example, NumPy is used to create an array of time points (time), as well as arrays representing 
# the contributions of two variables (variable1 and variable2). The np.cumsum() function is then used to 
#calculate the cumulative values (cumulative1 and cumulative2). Finally, Matplotlib's plt.stackplot() 
# function is employed to create the stack plot, and the plot is displayed with labels, legend, 
# and titles using Matplotlib functions.
```

# Sub-plot

```python
#  A subplot refers to a grid of smaller axes (plots) within a single figure. Subplots allow you to display 
# multiple plots in a single figure, making it easier to compare and analyze different aspects of your data.

import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating subplots
fig, axes = plt.subplots(2, 1, sharex=True)

# Plotting in subplots
axes[0].plot(x, y1, label='sin(x)')
axes[0].set_title('Sine Function')

axes[1].plot(x, y2, label='cos(x)', color='orange')
axes[1].set_title('Cosine Function')

# Adding legends and adjusting layout
for ax in axes:
    ax.legend()
plt.tight_layout()
plt.show()

# In this example, two subplots are created to visualize the sine and cosine functions in separate 
# panels within a single figure.
```
