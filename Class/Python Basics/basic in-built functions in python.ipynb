# Packing in python

```python
a,b,c=my_tup
print(a)
print(b)
print(c)
```

```python
my_tup1=56,7,7,5,7,5,7,5
a,b,c = my_tup1
print(a)
print(b)
print(c)
```

```python
my_tup1=56,7,7,5,7,5,7,5
a,b,c, *d = my_tup1
print(b, *d)
```

```python
my_tup1=56,7,7,5,7,5,7,5
a,b,c, *d = my_tup1
print(a)
print(*d)
```

# for loop - and the while loop. 

```python
# for loop is a countable loop.
# while loop is not a countable loop. it is infinite. 
# while loop is a continous loop and demands special commands for it to break.
for i in range(5):
    print("Hello, Gee")
```

```python
# while cond: this prints a continous output.
#   do_somthing
for i in range(5):
    while 0<1:
        print("Come early")
```
